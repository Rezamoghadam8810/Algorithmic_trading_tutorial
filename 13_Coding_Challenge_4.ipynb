```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('ggplot')
```

```python
close = pd.read_csv("/content/close.csv", index_col="Date", parse_dates=["Date"])
close
```

```python
BA=close.BA.dropna().to_frame().copy()
BA.rename(columns={'BA': 'Price'}, inplace=True)
BA
```

```python
BA.shift(periods=1)
```

```python
BA["P_lag1"]=BA.shift(periods=1)
BA
```

```python
BA["P_diff"]=BA.Price.sub(BA.P_lag1)
```

```python
BA
```

```python
BA["P_diff2"]=BA.Price.diff(periods=1)
```

```python
BA
```

```python
BA.P_diff.equals(BA.P_diff2)
```

```python
BA.Price.div(BA.P_lag1)-1
```

```python
BA["Returns"]=BA.Price.pct_change(periods=1)
BA
```

```python
BA.drop(columns=["P_lag1", "P_diff","P_diff2"], inplace=True)
```

```python
BA
```

```python
BA['log_ret']=np.log(BA.Price / BA.Price.shift())
BA
```

```python
BA.describe()
```

```python
mu = BA.log_ret.mean()
mu
```

```python
sigma=BA.log_ret.std()
sigma
```

**Investment Multiple**

```python
BA.Returns.add(1).prod() # cpmpounding simple returns
```

```python
np.exp(BA.log_ret.sum()) # sdding log returns
```

**Normalized Prices with Base 1**

```python
BA.Returns.add(1).cumprod() # compounding simple returns
```

```python
np.exp(BA.log_ret.cumsum()) # adding log returns
```

```python
BA.log_ret.cumsum().apply(np.exp) # adding log returns
```

**CAGR**

```python
(BA.Price[-1]/BA.Price[0])**(1/((BA.index[-1] - BA.index[0]).days / 365.25)) -1
```

```python
trading_days_year = BA.Returns.count() / ((BA.index[-1] - BA.index[0]).days / 365.25)
trading_days_year
```

```python
np.exp(BA.log_ret.mean() * trading_days_year) - 1 # correct with mean of dayly log returns
```
